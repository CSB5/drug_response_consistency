```python
def plot_pair(t0,t,t2,i,shapename,dataname):
    
    if (dataname=="GDSC"):
        noe=9
        #print(len(set(cell_line_details.values[:,1] )), len(set(cl_list)))
        
    else:
        noe=8
    
    plt.figure()
    
    plt.title(str(t0.values[i,0])+" : "+str(t0.values[i,1])+"\n"+str(t.values[i,0:noe].astype(int)))
    plt.scatter(t2.values[i,0:noe],t.values[i,0:noe])
    plt.ylim(-100,50)
    plt.xlabel("log2(Drug Dosage)")
    plt.ylabel("Cell viability")
    plt.savefig("../figures/Outlier_Curves/"+dataname+"_Shape_"+shapename+"/"+str(t0.values[i,0]) + "-" +str(t0.values[i,1])+".png")
    #plt.show()
```

```python
clnames=names.values[:,0]
drnames=names.values[:,1]
cl_set=(list(sorted(set( clnames ) )))
dr_set=(list(sorted(set( drnames ) )))
```

```python
(cell_line_tissue.values[:,1])

tissueType=[]
for i in range(len(cl_set)):
    ind=np.where(cell_line_tissue.values[:,0]==cl_set[i] )[0]
    
    tissueType.append( cell_line_tissue.values[ ind , 1 ][0] )
```

```python
pd.DataFrame(tissueType).to_csv('../data_processed/CurveCategoryDetection/TissueTypePerCL.csv', index=False)
```

```python
ti_set=list(sorted( set(tissueType) ) )
```

```python

flag_1=[0]*len(t)
flag_2=[0]*len(t)
flag_3=[0]*len(t)
flag_4=[0]*len(t)

category_matrix= np.zeros(shape=(len(cl_set),len(dr_set)))

drug_map=pd.DataFrame()
cl_map=pd.DataFrame()
ti_map=pd.DataFrame()

drug_map['DRUG_NAME'] = dr_set
cl_map['CL_NAME'] = cl_set
ti_map['TISSUE_NAME'] = ti_set #tissue map

drug_map['OUTLIER_COUNT'] = [0]*len(dr_set)
cl_map['OUTLIER_COUNT'] = [0]*len(cl_set)
ti_map['OUTLIER_COUNT'] = [0]*len(ti_set)

drug_map['#cl_tested'] = [0]*len(dr_set)
cl_map['#dr_tested'] = [0]*len(cl_set)
ti_map['#dr_tested'] = [0]*len(ti_set)
```

```python
response_range=200#np.nanpercentile(t.values[:,0], 99.99) - np.nanpercentile(t.values[:,-1], 0.01)
```

###### commented out assigning DW and DNW category  bcz we just want to show outliers (Hump/rising) in heatmap (0:Not tested,1: Tetsed, 3:hump, 4:rising)

```python
for i in range(t.shape[0]):
    print(i)
    
    sample=t.values[i,:]
    sample_sorted=sorted(list(sample))
    
    m=sample[round(len(sample)/2)]
    diff=sample[-1]-sample[0]
    ht_normalized=max(sample)-min(sample)
    ht=max(sample)-sample[0]
    ht2=sample_sorted[-2]-sample[0]
    
    category_matrix[cl_set.index(names.values[i,0]),dr_set.index(names.values[i,1])]=1
    
    drug_id=dr_set.index(names.values[i,1])
    cl_id=cl_set.index(names.values[i,0])
    
      
    ind=np.where(cell_line_tissue.values[:,0]==names.values[i,0] )[0]
    #print(names.values[i,0],ind,cell_line_tissue.values[ ind , 1 ])
    ti_id=ti_set.index( cell_line_tissue.values[ ind , 1 ]  )
    
    dr_freq=len(names[ names["Drug _Name"]== names.values[i,1] ])
    cl_freq=len(names[ names["Cell_line_name"]== names.values[i,0] ])
    
    drug_map.set_value( drug_id ,'#cl_tested', dr_freq)
    cl_map.set_value( cl_id ,'#dr_tested', cl_freq)
    ti_map.set_value( ti_id ,'#dr_tested', cl_freq+ti_map.values[ti_id,2])
 

    #print(str(t0.values[i,0])+" : "+str(t0.values[i,1]))
    if(abs(diff)<5 and ht>0.15*response_range and ht2>0.1*response_range):#5*diff):
        
        flag_3[i]=1
        category_matrix[cl_id,drug_id]=3
        plot_pair(names, response, logDosage, i, "Hump", dataname)
        
        drug_map['OUTLIER_COUNT']
        
        
        drug_map.set_value( drug_id ,'OUTLIER_COUNT', 1+drug_map.values[drug_id,1])
        cl_map.set_value( cl_id ,'OUTLIER_COUNT', 1+cl_map.values[cl_id,1])
        ti_map.set_value( ti_id ,'OUTLIER_COUNT', 1+ti_map.values[ti_id,1])
        
    elif(diff>0.15*response_range):#(diff>0.2*ht):
        flag_4[i]=1
        category_matrix[cl_id,drug_id]=4
        plot_pair(names, response, logDosage, i, "Rising", dataname)
        
        drug_map.set_value( drug_id ,'OUTLIER_COUNT', 1+drug_map.values[drug_id,1])
        cl_map.set_value( cl_id ,'OUTLIER_COUNT', 1+cl_map.values[cl_id,1])
        ti_map.set_value( ti_id ,'OUTLIER_COUNT', 1+ti_map.values[ti_id,1])
        
    elif(m-sample[0] <-0.025*response_range and sample[-1]-m<-0.025*response_range):
        flag_2[i]=1
        #category_matrix[cl_id,drug_id]=2 
        #plot_pair(names, response, logDosage, i, "S", dataname)
    else:
        flag_1[i]=1
        #category_matrix[cl_id,drug_id]=1
        #plot_pair(names, response, logDosage, i, "Constant", dataname)
    
```
